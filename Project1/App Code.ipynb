```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
def read_file(f):
    size_array = f.readline().split()
    d2_array = np.genfromtxt(f, dtype=str, delimiter='\t')
    return d2_array
```

```python
def plot_graph(d2_array):
    print(
    '''You can do a plot of any two features of the Iris Data set.
    The feature codes are: 
    0  =  sepal length
    1  =  sepal width
    2  =  petal length
    3  =  petal width
    ''')
    while True:
        hor = int(input('Enter feature code for the horizontal axis:'))
        ver = int(input('Enter feature code for the vertical axis:'))
        
        try:
            check_hor = d2_array[:, hor].astype(float)
            check_ver = d2_array[:, ver].astype(float)
            break
        except (IndexError, ValueError):
            print('Wrong values entered, please try again.(Only accepts values beetween 0 to 3')
    
    fig, ax = plt.subplots()
    for row in d2_array:
        if row[4] == 'Setosa':
            ax.scatter(float(row[hor]), float(row[ver]), c = 'green', marker = 'v')
        elif row[4] == 'Versicolor':
            ax.scatter(float(row[hor]), float(row[ver]), c = 'blue', marker = 'o')   
        elif row[4] == 'Virginica':
            ax.scatter(float(row[hor]), float(row[ver]), c = 'red', marker = '*')

    legend_elements = [Line2D([0], [0], marker='v', color='w', markerfacecolor='g', label='Setosa', markersize=10),
               Line2D([0], [0], marker='o', color='w', markerfacecolor='b', label='Versicolor', markersize=10),
               Line2D([0], [0], marker='*', color='w', markerfacecolor='r', label='Virginica', markersize=15)
               ]
    plt.legend(handles=legend_elements)
    plt.xlabel(flower_dict[hor])
    plt.ylabel(flower_dict[ver])
    plt.title('Iris Flower Pot')
    
    plt.show()      
```
